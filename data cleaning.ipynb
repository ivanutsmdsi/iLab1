```python
## import libraries
import pandas as pd
import glob
import numpy as np
from nltk.corpus import stopwords
```

# Read and merge output files

```python
# load all output files and merge to single dataframe
files = glob.glob("scrapper_output\*.csv")

# save to panda DF
df = pd.concat(map(pd.read_csv, glob.glob("scrapper_output\*.csv")))

# show sample
df.head(3)
```

# Clean plos abstract before merging text for analysis

```python
# get plos:abstract subset
plos_data = df['scrapper'] == 'plos_scrapper.py'
news_data = df['scrapper'] == 'news_scrapper_2.py'

df_plos_data = df[plos_data]

# list results
df_plos_data.head(3)
```

```python
# remove the first 2 and last 2 characters from PLOS abstract, else return abstract in full
df['abstract'] = np.where(df['scrapper'] == 'plos_scrapper.py', df['abstract'].str[2:-2]  , df['abstract'] )
df[plos_data].head(3)
```

```python
# remove the the \n character
df['abstract'] = np.where(df['scrapper'] == 'plos_scrapper.py', df['abstract'].str.replace(r'\\n', '')  , df['abstract'] )
df[plos_data].head(3)
```

# Fix date column

Plos date format is 2022-08-04T00:00:00Z, %Y-%M-%dT%h:%m:%sZ

google news date format is Wed, 10 Aug 2022 07:00:00 GMT

```python
from datetime import datetime

plos_date = "2022-08-04T00:00:00Z"
google_date = "Wed, 10 Aug 2022 07:00:00 GMT"

p_format = "%Y-%m-%dT%H:%M:%SZ"
g_format = "%a, %d %b %Y %H:%M:%S GMT"

date = datetime.strptime(plos_date, p_format)
date2 = datetime.strptime(google_date, g_format)

print(date)
print(date2)
```

```python
from datetime import datetime
p_format = "%Y-%m-%dT%H:%M:%SZ"
g_format = "%a, %d %b %Y %H:%M:%S GMT"

df['published'] = pd.to_datetime(df['published'], format = p_format, errors = "coerce").fillna(pd.to_datetime(df['published'], format = g_format, errors = "coerce"))

df.head(3)
```

# Identify columns of interest in data set based on [type]

```python
# build case logic
df['text'] = np.where(df['type'].str.contains(':abstract'), df['abstract'],
             np.where(df['type'].str.contains(':title'), df['title'],
             None))

# remove entries if no text was collected - means somethings went wrong in data collection
df = df.dropna(subset = ['text'])

df.head(3)
```

```python
text_df = df[['published', 'text', 'label', "type"]].copy()

text_df.head(5)
```

## Combined Text Cleaning Steps
1. text to lower case
2. remove non-english characeters, punctuation, numbers and whitespaces
3. remove stopwords
4. stemming words

```python
# apply lowercase

text_df['text'] = text_df['text'].str.lower()
```

```python
# Remove non-english characeters, punctuation, numbers and whitespaces

import re

## ([^A-Za-z- \t]) <- keep alphabetical characters and hyphens
## ( - )   <- discard hyphen when it is on its own
text_df['text2'] = text_df['text'].apply(lambda x: re.sub(r"([^A-Za-z- \t])|( - )", "", x))

## clean up extra whitespaces if present
# "\s\s+" <- looks for trailing whitespaces
text_df['text2'] = text_df['text2'].apply(lambda x: re.sub(r"\s\s+", " ", x))

#trim whitespaces at edges
text_df['text2'] = text_df['text2'].str.strip()
```

```python
# Remove stopwords
# import nltk                <- only necessary if library not yet downloaded
# nltk.download('stopwords') <- only necessary if library not yet downloaded
from nltk.corpus import stopwords

stop_words = stopwords.words('english')

text_df['text3'] = text_df['text2'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
```

```python
# stemming
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()

text_df['text4'] = text_df['text3'].apply(lambda x: [stemmer.stem(word) for word in x.split()])
```
